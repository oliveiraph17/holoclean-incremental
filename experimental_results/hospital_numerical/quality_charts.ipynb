```python
import matplotlib.pyplot as plt
import pandas as pd

from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))

########################################
# DataFrames - Setup
########################################

df = pd.read_csv('co_c_quality_log.csv', sep=';')
df = df[['infer_mode', 'skip_training_starting_batch', 'batch', 'precision', 'recall', 'f1']]

header_rows_df = df.loc[df['infer_mode'] == 'infer_mode']
header_row_numbers = header_rows_df.index.to_list()

df_list = []
beginning = 0
for row_number in header_row_numbers:
    end = row_number - 1

    tmp_df = df.loc[beginning:end]
    tmp_df = tmp_df.astype({'batch': 'int32', 'precision': 'float64', 'recall': 'float64', 'f1': 'float64'})

    df_list.append(tmp_df)
    beginning = row_number + 1
tmp_df = df.loc[beginning:]
tmp_df = tmp_df.astype({'batch': 'int32', 'precision': 'float64', 'recall': 'float64', 'f1': 'float64'})
df_list.append(tmp_df)
```

```python
########################################
# Charts - Setup
########################################

plt.rcParams.update({'font.size': 18})
plt.style.use('seaborn-darkgrid')
palette = plt.get_cmap('tab10')
```

```python
########################################
# Charts - DK 2%
########################################

count = 0

chart_rows = 16
chart_columns = 3
fig, axes = plt.subplots(chart_rows, chart_columns, sharex=True, sharey=True, figsize=(chart_columns * 7, chart_rows * 5))

for i in range(chart_rows):
    for j in range(chart_columns):
        current_df = df_list[count]
        count += 1

        axes[i, j].plot('batch', 'precision',
                        data=current_df, marker='', color=palette(1), label='precision', linewidth=3)
        axes[i, j].plot('batch', 'recall',
                        data=current_df, marker='', color=palette(2), label='recall', linewidth=3)
        axes[i, j].plot('batch', 'f1',
                        data=current_df, marker='', color=palette(3), label='f1', linewidth=3)

        skip_training_starting_batch = current_df['skip_training_starting_batch'].iloc[0]
        title_str = 'Skips training from batch {} onwards'.format(skip_training_starting_batch)
        axes[i, j].set(title=title_str)
        
        if j == 0:
            axes[i, j].set(ylabel='Percentage')
        if i == chart_rows - 1:
            axes[i, j].set(xlabel='Batch')

axes[chart_rows - 1, chart_columns - 1].legend(loc="lower left")
fig.tight_layout()
```
